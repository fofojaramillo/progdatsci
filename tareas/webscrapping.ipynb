```python
initial_link = 'https://icd.who.int/es'
home = 'https://icd.who.int'
```

```python
r = requests.get(initial_link)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
#Se obtienen todos los links de la pagina y clasifican por links a otras paginas y archivos .csv
def links_class(soup):
    files = []
    new_links = []
    links = soup.find_all('a')
    for link in links:
        href = link.get('href')
        if href and (href.endswith('.csv') or href.endswith('.zip')):
            if not href.startswith('https'):
                href = home + href
            files.append(href)
        if href and href.startswith('https://icd'):
            new_links.append(href)
    return files, new_links
```

```python
def pages_explorer(files, links, banned_sites):
    links_visited = []
    for link in links:
        banned = any(site in link for site in banned_sites)
        if link not in links_visited and not banned: 
            time.sleep(0.002)
            r = requests.get(link)
            links_visited.append(link)
            soup = BeautifulSoup(r.content, 'html.parser')
            new_files, new_links = links_class(soup)
        files.extend(new_files)
        links.extend(new_links)
    return files, links
```

```python
files, links = links_class(soup)
banned_sites = [
    'twitter',
    'x',
    'facebook',
    'instagram'
]

pages_explorer(files, links, banned_sites)
```

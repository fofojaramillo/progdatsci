```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
plt.style.available
```

```python
plt.style.use('bmh')
```

```python
x = np.linspace(0,10,100)
fig = plt.figure()
plt.plot(x,np.sin(x),'-')
plt.plot(x,np.cos(x),'--')
```

```python
fig.savefig('sencos.png', dpi = 100)
```

```python
fig, ax = plt.subplots(2,1)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```
